```python
from datasets import load_dataset

ds = load_dataset("BrightData/Goodreads-Books")
df = ds["train"].to_pandas()
```

```python
df
```

```python
df = df[df["genres"].notna() & (df["genres"] != "[]")]
```

```python
df=df[df["num_ratings"].notna() & df["num_reviews"].notna()]
```

```python
df=df[df["num_ratings"]>7500]
```

```python
df
```

```python
import ast
from collections import Counter
import pandas as pd

# convert string → list
#df["genres"] = df["genres"].apply(ast.literal_eval)

# count genres
genre_counts = Counter(
    genre
    for genres_list in df["genres"]
    for genre in genres_list
)

# convert to DataFrame (nice format)
genre_df = pd.DataFrame(genre_counts.items(), columns=["genre", "count"])
genre_df = genre_df.sort_values(by="count", ascending=False)

# print all
print(genre_df)

# optional: total number of distinct genres
print("\nNumber of distinct genres:", len(genre_df[genre_df["count"]>2000]))


genre_df.head(29)
```
